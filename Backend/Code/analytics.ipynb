```python
import pandas as pd
```

```python
# Store results in results dataframe
data = {'school' : ['NUS', "NTU", "SMU", "SUSS", "SUTD", "SIT"]}
results = pd.DataFrame(data)

# Read and Clean All Uni Module Info
nus_data = pd.read_excel("../../Data/university_courses/nus_dsa_mods.xlsx")
nus_data = nus_data.rename(columns = {'mod_category' : 'mod_cat', 'mod_exam' : 'mod_finals'})

ntu_data = pd.read_csv("../../Data/university_courses/NTU_course_info.csv")
ntu_data['mod_desc'] = ntu_data['Course Aims'] + " " + ntu_data["Intended Learning Outcomes"] + " " + ntu_data["Course Contents"]
ntu_data['mod_dept'] = ntu_data['Module Code'].astype(str).str[0:2]
ntu_data = ntu_data.rename(columns = {'Module Code' : 'mod_code', 'Module Name' : 'mod_name', 'Department' : 'mod_cat', 'Module Level' : 'mod_level', 'Pre-requisites': 'mod_prereq', 'Project': 'mod_proj', 'Final Exam' : 'mod_finals', 'Link' : 'mod_link'})
ntu_data = ntu_data.dropna(how='all', axis='columns')

smu_data = pd.read_csv("../../Data/university_courses/SMU_course_info.csv", skiprows = [0])
smu_data = smu_data.rename(columns = {'Module Code' : 'mod_code', 'Module Name' : 'mod_name', 'Department' : 'mod_dept', 'Module Level' : 'mod_level', 'Project': 'mod_proj', 'Final Exam' : 'mod_finals', 'Link' : 'mod_link', 'Module Category' : 'mod_cat', 'Description' : 'mod_desc'})

suss_data = pd.read_csv("../../Data/university_courses/SUSS_course_info.csv")
suss_data = suss_data.rename(columns = {'module code' : 'mod_code', 'module name' : 'mod_name', 'faculty' : 'mod_dept', 'level' : 'mod_level', 'prerequisites': 'mod_prereq', 'project ': 'mod_proj', 'final' : 'mod_finals', 'Link' : 'mod_link', 'module category' : 'mod_cat'})
cols = ['mod_finals', 'mod_proj']
suss_data[cols] = suss_data[cols].replace('T',1)
suss_data[cols] = suss_data[cols].replace('F',0)

sutd_data = pd.read_csv("../../Data/university_courses/SUTD_course_info.csv", skiprows = [0,1,2,3,4])
sutd_data['mod_desc'] = sutd_data['Module description'] + " " + sutd_data["Learning Objective"] + " " + sutd_data["Measurable Outcomes"]
sutd_data = sutd_data.rename(columns = {'Module code' : 'mod_code', 'Module Title' : 'mod_name', 'Department' : 'mod_dept', 'Prerequisites': 'mod_prereq', 'Project.1': 'mod_proj', 'Final.1' : 'mod_finals', 'Link' : 'mod_link', 'Module Category' : 'mod_cat', 'Module level' : 'mod_level', 'Preclusion' : 'mod_preclu'})
sutd_data = sutd_data.drop(columns = ['Final', 'Project'])

sit_data = pd.read_csv("../../Data/university_courses/SIT_Module_Info.csv")
sit_data['mod_desc'] = sit_data['Module Description '] + " " + sit_data["Topics"]
sit_data = sit_data.rename(columns = {'Module Code' : 'mod_code', 'Module Title' : 'mod_name', 'Department ' : 'mod_dept', 'Project.1': 'mod_proj', 'Final.1' : 'mod_finals', 'Link' : 'mod_link', 'Module Category' : 'mod_cat', 'Module Level' : 'mod_level', 'Department' : 'mod_dept'})

all_data = [nus_data, ntu_data, smu_data, suss_data, sutd_data, sit_data]
```

# Number of Modules Provided by School

Count Num of Rows 

```python
num_of_mods = []
for d in all_data:
    num_of_mods.append(len(d))
results['num_of_mods'] = num_of_mods
results
```

# Breakdown of Modules Offered By Departments

Group By Module Departments, Count Number of Modules

```python
mod_depts = []
for df in all_data:
    mod_depts.append(df['mod_dept'].value_counts().to_dict())
results['mod_depts'] = mod_depts
results
```

# Breakdown of Module Levels

Group by module levels, count

```python
mod_levels = []
for df in all_data:
    mod_levels.append(df['mod_level'].value_counts().to_dict())
results['mod_levels'] = mod_levels
results
```

# Overall Breakdown of Module Project & Finals

weighted proj-finals breakdown for each module e.g. for a module, proj and finals = 1, weighted proj and finals = 0.5 each

for a school, overall proj proportion = (sum of weighted proj score) / (sum of weighted proj score + sum of weighted finals score)

vice versa for overall finals proportion

```python
mod_proj_finals = []
for df in all_data:
    if 'mod_finals' and 'mod_proj' in df.columns:
        df.loc[(df['mod_finals'] == 1) & (df['mod_proj'] == 1), ['mod_finals', 'mod_proj']] = 0.5
        num_of_finals = df['mod_finals'].sum()
        num_of_proj = df['mod_proj'].sum()
        total_num = num_of_finals + num_of_proj
        mod_proj_finals.append({'finals' : num_of_finals / total_num * 100, 'proj' : num_of_proj / total_num * 100})
    else:
        mod_proj_finals.append({})
results['mod_proj_finals'] = mod_proj_finals
results
```

# Breakdown of Module Categories

```python
mod_cats = []
for df in all_data:
    old = df['mod_cat'].value_counts().to_dict()
    new = {}
    for key in list(old.keys()):
        if ',' in key:
            cats = list(key.split(', '))
            for c in cats:
                if c in new:
                    new[c] += 0.5 * old[key]
                else:
                    new[c] = 0.5 * old[key]
        else:
            if key in new:
                new[key] += old[key]
            else:
                new[key] = old[key]

    mod_cats.append(new)
    
results['mod_cats'] = mod_cats
results
```

# Breakdown of Modules in Each Category

```python
mods_in_cat = []
for df in all_data:
    df = df.groupby(['mod_cat'])['mod_name'].apply(list).reset_index(name = 'mod_names_in_cat')
    old = df.set_index('mod_cat').T.to_dict('records')[0]
    new = {}
    for key in list(old.keys()):
        if ',' in key:
            cats = list(key.split(', '))
            for c in cats:
                if c in new:
                    new[c].extend(old[key])
                else:
                    new[c] = old[key]
        else:
            if key in new:
                new[key].extend(old[key])
            else:
                new[key] = old[key]

    mods_in_cat.append(new)
results['mods_in_cat'] = mods_in_cat
results
```

# Breakdown of Module Proj & Finals in Each Category

Group By Category
Compute Proj & Finals Proportion

```python
proj_finals_in_cat = []
for df in all_data:
    if 'mod_finals' and 'mod_proj' in df.columns:
        df.loc[(df['mod_finals'] == 1) & (df['mod_proj'] == 1), ['mod_finals', 'mod_proj']] = 0.5
        df = df.groupby('mod_cat').agg({'mod_proj': 'sum', 'mod_finals' : 'sum'}).reset_index()
        old = df.set_index('mod_cat').T.to_dict()
        new = {}
        for key in list(old.keys()):
            if ',' in key:
                cats = list(key.split(', '))
                for c in cats:
                    if c in new:
                        new[c]['mod_proj'] += 0.5 * old[key]['mod_proj']
                        new[c]['mod_finals'] += 0.5 * old[key]['mod_finals']
                    else:
                        new[c] = {'mod_proj' : 0.5 * old[key]['mod_proj'], 'mod_finals' : 0.5 * old[key]['mod_finals']}
            else:
                if key in new:
                    new[key]['mod_proj'] += old[key]['mod_proj']
                    new[key]['mod_finals'] += old[key]['mod_finals']
                else:
                    new[key] = {'mod_proj' : old[key]['mod_proj'], 'mod_finals' : old[key]['mod_finals']}

    proj_finals_in_cat.append(new)
results['proj_finals_in_cat'] = proj_finals_in_cat
results
```

# Foundation of Each Category

Group By Category and Count mean(Number of each Pre-req of a module)? nus, ntu, suss, sutd

```python
found_of_cat = []
for df in all_data:
    df = df.groupby(['mod_cat'])['mod_name'].apply(list).reset_index(name = 'mod_names_in_cat')
    old = df.set_index('mod_cat').T.to_dict('records')[0]
    new = {}
    for key in list(old.keys()):
        if ',' in key:
            cats = list(key.split(', '))
            for c in cats:
                if c in new:
                    new[c].extend(old[key])
                else:
                    new[c] = old[key]
        else:
            if key in new:
                new[key].extend(old[key])
            else:
                new[key] = old[key]

    found_of_cat.append(new)
results['found_of_cat'] = found_of_cat
results
```
